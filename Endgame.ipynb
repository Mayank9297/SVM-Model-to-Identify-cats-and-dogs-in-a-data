```python
input_dir = r"C:\Users\mayan\Desktop\Project\Project two\cat_dog_images"
categories = ['Cats','Dogs']
data = []
labels = []
for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir, category)):
        img_path = os.path.join(input_dir, category, file)
        print(img_path)
        img = imread(img_path)
        img = resize(img, (15,15))
        data.append(img.flatten())
        labels.append(category_idx)
```

```python
data[1]
```

```python
labels[1]
```

Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels) 
```

Build SVM Model for Classification

```python
from sklearn.svm import SVC
svm_model = SVC()
```

Model Training

```python
svm_model.fit(x_train, y_train)
```

Model Prediction

```python
y_pred = svm_model.predict(x_test)
```

```python
y_pred
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
score
```

```python
def image_classification_prediction(image):
    img_new = imread(image)
    img_new1 = resize(img_new, (15,15))
    img_flatten = img_new1.flatten()
    img_array = np.asarray(img_flatten)
    result = svm_model.predict(img_array.reshape(1,-1))
    if result[0] == 1:
        print ("It's a cat")
    else:
        print("It's a Dog")
```

```python
import os
folder_path = "C:\\Users\\mayan\\Desktop\\Project\\Project two\\test1"
for filename in os.listdir(folder_path):
  if filename.endswith((".jpg", ".jpeg", ".png")):
    image_path = os.path.join(folder_path, filename)
    image_classification_prediction(image_path)




```

```python
import os

def get_filenames(folder_path, extension=None):
  filenames = []
  for filename in os.listdir(folder_path):
    if extension is None or filename.endswith(extension):
      filenames.append(filename)
  return filenames


folder_path = r"C:\Users\mayan\Desktop\Project\Project two\test1"
image_filenames = get_filenames(folder_path, extension=".jpg")  

print("Image filenames:", image_filenames)

```

```python
data = []
for filename, label, predicted_label in zip(image_filenames, y_test, y_pred):  
  data.append([filename, label, predicted_label])  

csv_output_path = "svc_results.csv"

with open(csv_output_path, 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)

  writer.writerow(['Filename', 'True Label', 'Predicted Label'])  # Adjust columns as needed

  writer.writerows(data)

print(f"CSV file created: {csv_output_path}")
```
